```python
import zipfile
```

```python
with zipfile.ZipFile("/content/flower-color-images.zip", "r") as zip_ref:
    zip_ref.extractall("/content/flower-color-images")
```

```python
directory_path = "/content/flower-color-images/flower_images/flower_images"
```

```python
file_list = os.listdir(directory_path)
```

```python
file_list = [file for file in file_list if os.path.isfile(os.path.join(directory_path, file))]
```

```python
image_path = os.path.join(directory_path, file_list[0])
```

```python
original_image = load_img(image_path)
```

```python
rotated_image = original_image.rotate(90)
```

```python
rotated_image1 = original_image.rotate(180)
```

```python
zoomed_image = original_image.resize((int(1.2*original_image.size[0]), int(1.2*original_image.size[1])))
```

```python
def show_images(original, rotated, title):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(rotated)
    plt.title(title)
    plt.axis('off')

    plt.show()
```

```python
show_images(original_image, rotated_image, 'Rotated Image')
```

```python
show_images(original_image, rotated_image1, 'Rotated Image1')
```

```python
show_images(original_image, zoomed_image, 'Zoomed Image')
```
